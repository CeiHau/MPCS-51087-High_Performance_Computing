```python
from PIL import Image
import pandas as pd
```

```python
img = pd.read_csv('G_6.csv', header = None)
img = img.drop(columns = 1000)
display(img.head(),)
img = img.to_numpy()
```

```python

print(img)
```

```python
img = img * 255
im = Image.fromarray(img)
im = im.convert('L')  # 这样才能转为灰度图，如果是彩色图则改L为‘RGB’
im.save('outfile.png')
```
