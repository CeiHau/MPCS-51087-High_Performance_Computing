## Strong Scaling

```python
import matplotlib.pyplot as plt
thread_num = [1, 2, 4, 8, 16, 32, 64]
T = [343,182.73, 77.83, 38.55, 19.74, 10.70, 5.29]

Sn = []
for i in T:
    Sn.append(T[0]/i)
plt.plot(thread_num,Sn)
plt.title('Strong scaling')
plt.xlabel('n')
plt.ylabel('Sn')
plt.show()
print(Sn)
```

## Weak Scaling

```python
import matplotlib.pyplot as plt
thread_num = [1,2, 4, 8, 16, 32, 64]
time = [1.92, 2.72, 3.78, 15.57, 7.49, 10.19, 15.76]
S = []
for i in time:
    S.append(time[0]/i)
plt.plot(thread_num,S)
plt.title('Weak scaling')
plt.xlabel('n')
plt.ylabel('S')
plt.show()
```
